# ImageMagick Montage Quick Guide

Make sure you have [ImageMagick installed](https://imagemagick.org/script/download.php).

images (CC BY-SA 3.0)
- https://commons.wikimedia.org/wiki/File:Baywing_(Agelaioides_badius).jpg
- https://commons.wikimedia.org/wiki/File:Sitta-carolinensis-001.jpg

```python
from IPython.display import Image
img = lambda n: Image(filename=n)
```

## basic montage

```python
!montage baywing.jpg sitta.jpg montages/basic.jpg
img('montages/basic.jpg')
```

## let's make the background transparent
The output image needs to be PNG to have transparent background.

```python
!montage -background none baywing.jpg sitta.jpg montages/nobg.png
img('montages/nobg.png')
```

## let's add some labels

```python
!montage -label Baywing baywing.jpg -label Sitta sitta.jpg montages/labels.jpg
img('montages/labels.jpg')
```

## let's title the montage!

```python
!montage -title 'Two Birds' baywing.jpg sitta.jpg montages/title.jpg
img('montages/title.jpg')
```

## resize
the montage is too small. let's increase the size.

```python
!montage -geometry 200x200 baywing.jpg sitta.jpg montages/resized.jpg
img('montages/resized.jpg')
```

## putting it all together

```python
!montage -geometry 200x200 -background none -title 'Two Birds' -label Baywing baywing.jpg -label Sitta sitta.jpg montages/all.png
img('montages/all.png')
```
